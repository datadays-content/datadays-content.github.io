<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
استنتاج آماری
        </font>
		<p></p>
		<hr>
        کلیت روش‌های مربوط به فرضیه بر اساس گامهای زیر است:
        <ol>
            <li>
                در نظر گرفتن یک فرض اولیه. 
            </li>
            <li>
                جمع آوری شواهد(داده‌ها)
            </li>            
            <li>
                با توجه به داده‌های جمع‌آوری شده، تصمیم می‌گیریم که فرض اولیه را رد کنیم یا خیر. 	
            </li>
            </ol>
        <br>
        شش بخش آزمودن یک فرضیه به شرح زیر است:
        <ol>
            <li>
                داشتن دو فرضیه که در تقابل یک‌دیگر هستند. 
            </li>
            <li>
                معین کردن سطح اهمیت یا خطای نوع اول(که بعدا مشخص می‌شود کدام خطا است) که معمولا آن را &alpha; می‌نامیم.
            </li>            
            <li>
                محاسبه‌ی یک آماره‌ی متناظر با فرصیه.
            </li>
            <li>
                محاسبه‌ی احتمال وقوع(<i>p-value</i>) و تعیین مرز برای بازه‌های مردود بودن فرضیه.
            </li>
            <li>
                در نظر گرفتن یک آزمون صحت یا عدم صحت برای فرض صفر (همان فرض اولیه‌ی مد نظرمان)
            </li>
            <li>
                بیان نتیجه‌گیری کلی حاصل از تست
            </ol>
<br>
</font>
</div>

<div id="sec_hypo_test" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
آزمون فرضیه
</font>
		<p></p>
		<hr>
        عناصر اصلی آزمون فرضیه عبارت‌اند از:
        <ul>
            <li>
                <b>فرض صفر(H<sub>0</sub>) و فرض مقابل آن(H<sub>1</sub>):</b> فرض صفر در واقع همان فرضیه‌ای است که می‌خواهیم آن را مورد بررسی قرار دهیم. حالات مختلف فرض مقابل را در قسمت بعد توضیح خواهیم داد
            </li>
            <li>
                <b>آماره‌ی آزمون:</b> آن آماره‌ی خاصی از داده‌ها که بر اساس آن تصمیم می‌گیریم تا H<sub>0</sub> (فرض صفر) را رد کنیم و یا قبول کنیم.
            </li>
            <li>
                <b>مقادیر بحرانی:</b> مقادیری از آماره‌ی آزمون که مرز بین ناحیه‌ی مردودی فرضیه و ناحیه‌ی قابل پذیرش بودن فرضیه را مشخص می‌کند.
            </li>
            <li>
                <b>ناحیه‌ی مردودی:</b> مجموعه مقادیری از آماره‌ی آزمون که به ازای آنها نتیجه می‌گیریم که فرضیه مردود بوده است.
            </li>
            <li>
                <b>ناحیه‌ی پذیرش:</b> مجموعه مقادیری از آماره‌ی آزمون که به ازای آنها نتیجه می‌گیریم که فرضیه با دقت مورد نظر مردود نبوده است.
            </li>
            <li>
                <b>P-value یا probability value:</b> احتمال اینکه آماره‌ی آزمون به شرط صحت فرض صفر، حداقل به شدت مقدار مشاهده شده باشد.
        </ul>
        در کل دو نوع خطا می‌توانیم داشته باشیم:

  <tr>
    <th>Decision</th>

  </tr>

  </tr>

<div id="sec_null_hypo" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
فرض صفر و انواع فرض مقابل آن
</font>
		<p></p>
		<hr>
        همان طور که قبلا گفتیم، فرض صفر در واقع همان فرضیه‌ای است که می‌خواهیم آن را مورد بررسی قرار دهیم. مثلا مجموعه‌ای به اندازه‌ی ۱۰۰ نمونه از اطلاعات قد مردان کشور گینه بیسائو به ما داده شده است و میانگین قدشان 174 سانتی‌متر است. فرض صفر ما می‌تواند این باشد که میانگین قد کل مردان این کشور 170 سانتی‌متر است. یا مثلا مجموعه نمرات ریاضی ۲۵ دانش‌آموز پسر و ۳۰ دانش‌آموز دختر به ما داده شده است و میانگین پسرها و دخترها به ترتیب 14.5 و 15 است. فرض صفر را می‌توانیم تساوی میانگین نمره‌ی پسرها و دخترها در نظر بگیریم.
        <br>
        فرض مقابل فرض صفر را به سه شکل می‌توانیم در نظر بگیریم:
        <ul>
            <li>
                فرض مقابل این باشد که پارامتر جامعه‌ی آماری با مقدار معینی نامساوی است. در این حالت به آزمون، <b>آزمون دوطرفه</b> گفته می‌شود. <i>H<sub>1</sub>: p &ne; p<sub>0</sub> or &mu; &ne; &mu;<sub>0</sub></i><br>
مثلا در مورد همان نمره‌ی ریاضی فرض صفر، تساوی میانگین دو گروه، و فرض مقابل آن عدم تساوی میانگین است.            
            
  </li>
            <li>
                فرض مقابل این باشد که پارامتر جامعه‌ی آماری از مقدار معینی بیشتر است. در این حالت به آزمون، <b>آزمون یک‌طرفه از راست</b> گفته می‌شود. <i>H<sub>1</sub>: p &gt; p<sub>0</sub> or &mu; &gt; &mu;<sub>0</sub></i><br>
    مثلا می‌خواهیم بررسی کنیم که یک پیتزافروشی آیا به طور میانگین ظرف ۳۰ دقیقه غذا را آماده می‌کند؟(فرض صفر این می‌شود که میانگین مدت زمان لازم برای آماده شدن پیتزا، ۳۰ دقیقه است.) اکنون باید ببینیم فرض مقابل را چه انتخاب کنیم. منطقی نیست که فرض مقابل را عدم تساوی میانگین با ۳۰ دقیقه باشد بلکه در این حالت بهتر است فرض مقابل را «بیشتر بودن میانگین زمان مورد نیاز از عدد ۳۰» بگیریم.
            </li>
            <li>
                فرض مقابل این باشد که پارامتر جامعه‌ی آماری از مقدار معینی کمتر است. در این حالت به آزمون، <b>آزمون یک‌طرفه از چپ</b> گفته می‌شود. <i>H<sub>1</sub>: p &lt; p<sub>0</sub> or &mu; &lt; &mu;<sub>0</sub></i>
        </ul>

</font>
</div>

<div id="sec_p_value" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
توضیح در مورد $p-value$
</font>
		<p></p>
		<hr>
        <ul>
            <li>
گینه بیسائو یا جمهوری دموکراتیک کنگو؟ مساله این است!
            </li>
        </ul>
مثال قد مردان گینه بیسائویی(!) را به خاطر بیاورید. فرض صفر این بود که میانگین قد برابر با ۱۷۰ سانتی‌متر باشد. میانگین داده‌هایی که ما داریم، ۱۷۴ سانتی‌متر است. اکنون می‌خواهیم $p-value$ را محاسبه کنیم. احتمال این که در یک داده‌ی نمونه‌ی دیگر از قد مردان گینه بیسائو میانگین قد افراد نمونه بیش از ۱۷۴ باشد، به شرطی که میانگین قد جامعه ۱۷۰ باشد(یعنی فرض صفر درست باشد)، همان مقدار $p-value$ است. <br>همین‌جا برای همیشه از گینه بیسائوی لوس و بی‌مزه خداحافظی می‌کنیم و مثال بهتری خواهیم زد.
        <br>
        <ul>
            <li>
پرتاب سکه
            </li>
        </ul>
        	داده‌های در دسترس، نتایج حاصل از ۱۰۰ پرتاب سکه (و حوصله‌ی فراوان یک انسان بیکار) است. فرض صفر این است که سکه عادی است (یعنی احتمال شیر و خط آمدن برابر است). فرض کنید تعداد پرتاب‌های شیر ۸۰ عدد شده‌باشد. $p-value$ احتمال این است که در آزمایش‌های دیگر تعداد پرتاب‌های شیر حداقل ۸۰ عدد باشد، در حالی که سکه عادی و منصفانه باشد.
  <br>
   سؤالی که پیش می‌آید این است که چه $p-value$ای مناسب است؟به یاد آورید که $\alpha$ به معنای درجه‌ی اهمیت است و قبل از انجام آزمون فرض مقدار آن را به دلخواه تعیین می‌کنیم. به صورت کلی $\alpha$ را ۵، ۱ یا ۰.۱ انتخاب می‌کنند. 
<br>
دقت کنید که $p-value$ احتمال است بنابراین همواره مقدار بین صفر تا ۱ می‌گیرد.
<br>
	به صورت زیر می‌توان دقیق‌تر مقادیر $p-value$ را تفسیر کرد:
    <br>
    <ol>
            <li>
$p-value$ بسیار کم (خیلی کمتر از $\alpha$) مدرکی قوی علیه صحت فرض صفر است. در این صورت فرض صفر را رد می‌کنیم.
            </li>
            <li>
$p-value$ بسیار بالا (خیلی بیشتر از $\alpha$) مدرکی ضعیف علیه صحت فرض صفر است. در این صورت موفق به رد کردن فرض صفر نمی‌شویم. 
            </li>
            <li>
$p-value$ای که بسیار نزدیک به $\alpha$ باشد را می‌توان به هر دو صورت تفسیر کرد. 
            </li>
        </ol>     
</font>
</div>

<div id="sec_mean_test" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
آزمون میانگین جامعه‌ی آماری و استفاده از توزیع t
</font>
		<p></p>
		<hr>
        فرض کنید که مدت زمان لازم برای آماده شدن پیتزا در یک رستوران را در طول یک روز یادداشت کرده‌اید. میانگین زمانهایی را که یادداشت کرده‌اید محاسبه می‌کنید و می‌بینید که برابر با ۴۰ شده است.
        <br>
        اکنون با خود فکر می‌کنید که آیا ممکن است میانگین زمان آماده شدن یک پیتزا ۳۰ دقیقه باشد؟ نکند من روز بدی را برای یادداشت کردن زمان لازم انتخاب کرده‌ام و کلا پیتزاها فقط همین یک روز ۱۰ دقیقه دیرتر آماده شده‌اند؟ به عبارتی این سوال پیش می‌آید که شاید هر اختلافی بین میانگین نمونه‌های داده با سایر زمانهای ممکن برای میانگین زمان مورد نیاز آماده شدن پیتزا، صرفا به خاطر واریانس نمونه برداری باشد؟
        <br>
        به طور کلی فرض صفر را این طور در نظر می‌گیریم که «هر اختلاف مشهودی، تصادفی و صرفا شانسی رخ داده است» در حالی که فرض مقابل آن، ردکننده‌ی این تصور است.
        <br>
        اگر واریانس جامعه برای ما معلوم باشد، بنابراین به راحتی می‌توانیم به این سوال پاسخ بدهیم.(در این صورت نیازی به توزیع t نیست و آماره‌ای که می‌سازیم، توزیع نرمال دارد.) اما در عمل واریانس جامعه برای ما معلوم نیست و به جای واریانس جامعه، صرفا از روی نمونه‌ها می‌توانیم $s^2$ را محاسبه کنیم که تخمینی از واریانس جامعه است:
        $$\overline{x} = \frac{1}{n}\sum_{i = 1}^n{x_i} \quad \quad
 s^2 = \frac{\sum_{i = 1}^n(x - \overline{x})^2}{n - 1}
                $$
   <br>
   مقدار $\overline{x}$ میانگین، $s^2$ واریانس و $n$ اندازه‌ی نمونه‌ی آماری است.
   <br><br>
   اکنون آماره‌ی مورد نظر خودمان را می‌سازیم:
   $$t = \frac{\overline{x} - \mu_0}{\sqrt{s^2/n}}$$
   این آماره، توزیع t دارد. توزیع t در واقع شبیه توزیع نرمال است و هر چقدر که پارامتر آن بزرگتر شود، به توزیع نرمال نزدیک‌تر می‌شود. پارامتر این توزیع در اینجا برابر با اندازه‌ی نمونه‌ی آماری منهای ۱ است. یعنی هر چقدر اندازه‌ی نمونه بزرگتر باشد، توزیع این آماره به نرمال نزدیک‌تر می‌شود.(چرا چنین انتظاری هم باید می‌داشتیم؟)
   <br>
   هر چقدر واریانس نمونه بزرگتر باشد، یعنی در داده‌ها پراکندگی بیشتری داشته‌ایم و در نتیجه اختلاف بین میانگین داده‌ها و فرض صفر قابل تحمل‌تر.
   <br>
   برای استفاده از این آزمون، متغیر تصادفی تحت آزمایش باید شرایط زیر را دارا باشد:
         <ol>
             <li>
                 باید پیوسته باشد.
             </li>
             <li>
                 باید توزیع تقریبا نرمال داشته باشد.  
             </li>
             <li>
                 نباید شامل داده‌های پرت باشد. 
             </li>
         </ol>
         <br>
   به طور کلی برای آزمون میانگین با استفاده از آزمون توزیع t  مراحل زیر را باید طی کرد:
   
<ol>
    <li>
        یک فرض اولیه برای آزمودن در نظر می‌گیریم. 
    </li>
    <li>
            سطح اطمینان (&alpha;) را مشخص می‌کنیم. یعنی می‌خواهیم خطای قابل قبول نوع اول را مشخص کنیم.
    </li>
    <li>
        مقادیر آماره‌های آزمون را محاسبه می‌کنیم:
        $$t = \frac{\overline{x} - \mu_0}{\sqrt{s^2/n}}$$
   و در مثال بررسی سالم بودن یک سکه که شکست یا پیروزی داریم، به شکل زیر محاسبه می‌کنیم:
            $$t = \frac{\overline{p}-p_{0}}{\sqrt{\frac{p_{0} (1-p_{0})}{n}}}$$
            
   <br>
    که در آن $\mu_0$ و $p_0$ همان مقادیر مربوط به فرض صفر هستند.
   </li>
   <li>
       مقادیر بحرانی مناسب را برای آزمون پیدا می‌کنیم.
    </li>
    <li>
        بررسی می‌کنیم که آیا مقدار محاسبه شده‌ی آماره در ناحیه‌ی مردودی قرار گرفته است یا خیر.
    </li>
    <li>
        نتیجه‌گیری را بیان می‌کنیم.
    </li>
    </ol>
</font>
</div>

<br>
<div id="sec_comp_test" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
آزمون مقایسه‌ی میانگین دو جامعه‌ی آماری
</font>
		<p></p>
		<hr>
برای آزمون مقایسه‌ی میانگین دو جامعه‌ی آماری، از فرمول زیر استفاده می‌کنیم:
$$t=\frac{{\overline{x}}_1-{\overline{x}}_2}{s_p \sqrt{\frac{1}{n_1}+\frac{1}{n_2}}}$$
        <br>که در آن $s_p$ از عبارت زیر محاسبه می‌شود:$$s_p =\sqrt{\frac{(n_1-1)s_1^2+(n_2-1)s_2^2}{n_1+n_2-2}}$$
        <br>
        کدهای زیر مثال از حالات مختلف این تست به ازای دو مجموعه‌ی داده‌ی فرضی x و y هستند:
</font>
</div>

```r
# Using the x; y interface:
t.test(x, y, alt = "two:sided") # H1: mean(x) != mean(y) (default)
t.test(x, y, alt = "less")      # H1: mean(x) < mean(y)
t.test(x, y, alt = "greater")   # H1: mean(x) > mean(y)
# Using the formula interface:
t.test(y ~ g, data = d, alt = "two:sided") # H1: level1 mean = level2 mean
t.test(y ~ g, data = d, alt = "less") # H1: level1 mean < level2 mean
t.test(y ~ g, data = d, alt = "greater") # H1: level1 mean > level2 mean
```

<div id="sec_fit_test" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
		<font color=#006FDE size=6>
آزمون خوبی برازیدن یک تابع توزیع بر یک جامعه‌ی آماری
</font>
		<p></p>
		<hr>
فرض کنید که تعدادی نمونه‌ی داده در اختیار داریم.
        <ul>
            <li>
                ابتدا فرض می‌کنیم که رفتار جامعه‌ی نمونه از فلان توزیع خاص پیروی می‌کند.
            </li>
            <li>
               فرض مقابل این است که این جامعه چنین توزیعی نداشته باشد.
            </li>
            <li>
                از یک آزمون مثل آزمون خوبی برازیدن مربع کای استفاده می‌کنیم.(Chi-square Goodness-of-fit)
            </li>
            <li>
                آماره‌ی مورد نظر را باید با بازه‌بندی برای متغیرهای پیوسته محاسبه کنیم. مثلا در مورد آزمون مربع کای، داده‌ها را به kبازه، بازه‌بندی می‌کنیم و درصد تکرار هر بازه را در جامعه‌ی نمونه محاسبه می‌کنیم.
                سپس آماره‌ی زیر را محاسبه می‌کنیم.
            </li>
            $$\chi^{2} = \sum_{i=1}^{k}(O_{i} - E_{i})^{2}/E_{i}$$
            
  <br>
      که در آن E مقدار احتمال آن بازه در تابع توزیع فرضیه و O احتمال وقوعی است که برای بازه بر حسب فراوانی آن بازه در جامعه‌ی آماری محاسبه کرده‌ایم. بدین ترتیب آماره‌ی این تست محاسبه می‌شود.
      <br><br>
      این آزمون در واقع غیرپارامتری بود. در ادامه به سایر آزمونهای غیرپارامتری می‌پردازیم.
</font>
</div>

<div id="sec_anova" style="direction:rtl;line-height:300%;">
	<font face="XB Niloofar" size=5>
        <ul>
            <li><b>ANOVA</b>
            </li>
            این آزمون شکل شکل کلی‌تر از آزمون t برای مقایسه میانگین بیش از 2 گروه می‌باشد.

            آزمون ANOVA نیز همانند t-test هم برای گروه‌های مستقل و هم برای وابسته (در هنگامی که چند مشاهده از یک متغیر داریم) می‌توان استفاده کرد.
        </ul>
      </font>
</div>

```r
anova()
```
